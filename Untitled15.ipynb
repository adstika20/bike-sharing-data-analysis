## **Data Cleaning**

```python
# Hanya menampilkan kolom baris yang mengandung missing value
df[df.ADDRESSLINE2.isna()]
```

```python
df.STATE.value_counts()
```

```python
# Menangani missing value dengan modus
most = df['STATE'].mode()[0]
# Mengisi nilai NaN pada kolom 'STATE' dengan modus
df['STATE'] = df['STATE'].fillna(most)
```

```python
df.STATE.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.ADDRESSLINE2.value_counts()
```

```python
mostt = df['ADDRESSLINE2'].mode()[0]

df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna(mostt)
```

```python
df.ADDRESSLINE2.isna().sum()
```

```python
df.TERRITORY.value_counts()
```

```python
mosttt = df['TERRITORY'].mode()[0]
df['TERRITORY']= df['TERRITORY'].fillna(mosttt)
```

```python
df.drop(columns='POSTALCODE', inplace=True)
```

```python
df.isna().sum()
```

# **EDA**

```python
# Ubah tipe data

df['ORDERDATE']=pd.to_datetime(df['ORDERDATE'])
```

```python
df.corr(numeric_only=True)
```

### JAWABAN 1

- SALES: Total penjualan.
- QTR_ID: Kuartal tahun.
- MONTH_ID: Bulan pesanan.
- YEAR_ID: Tahun pesanan.

```python
# sales, qtr, month, year
df.groupby(by='SALES').agg({
    'ORDERNUMBER': "nunique",
    'SALES':['max', 'min','mean','std']
})
```

```python
# tipe data
df.YEAR_ID.dtypes
```

```python
# Total penjualan per tahun

sales_year = df.groupby('YEAR_ID')['SALES'].sum()
print(sales_year)
```

```python
# total penjualan per kuartal

kuartal_year = df.groupby('QTR_ID')['SALES'].sum()
print(kuartal_year)
```

```python
# total penjualan per bulan

month_year = df.groupby('MONTH_ID')['SALES'].sum()
print(month_year)
```

```python
df['YEAR_ID'] = df['YEAR_ID'].astype(int)
```

# **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.barplot(x=sales_year.index.astype(int), y=sales_year.values)
plt.title('Total Penjualan per Tahun')
plt.xlabel('tahun')
plt.ylabel('total penjualan')
#plt.xticks(sales_year.index.astype(int))
plt.show()
```

```python
# penjualan per kuartal

plt.figure(figsize=(10,5))
sns.lineplot(x=kuartal_year.index, y=kuartal_year.values)
plt.title('total penjualan per kuartal')
plt.xlabel('kuartal')
plt.ylabel('total penjualan')
plt.show()
```

```python
# Plot penjualan per bulan
plt.figure(figsize=(12, 6))
sns.barplot(x=month_year.index, y=month_year.values)
plt.title('Total Penjualan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Penjualan')
plt.show()
```
